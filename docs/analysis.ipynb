```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Arrays con los tiempos
tiempos_1000 = np.array([0.049, 0.046, 0.060, 0.038, 0.037])
tiempos_2000 = np.array([0.095, 0.104, 0.116, 0.092, 0.156])
tiempos_4000 = np.array([0.176, 0.181, 0.197, 0.147, 0.213])
tiempos_8000 = np.array([0.349, 0.368, 0.358, 0.378, 0.372])
tiempos_16000 = np.array([0.636, 0.638, 0.605, 0.571, 0.628])
tiempos_25000 = np.array([0.655, 0.898, 0.670, 0.633, 0.690])

# Calcular el promedio de cada array
mean_1000 = np.mean(tiempos_1000)
mean_2000 = np.mean(tiempos_2000)
mean_4000 = np.mean(tiempos_4000)
mean_8000 = np.mean(tiempos_8000)
mean_16000 = np.mean(tiempos_16000)
mean_25000 = np.mean(tiempos_25000)

# Crear un DataFrame con los resultados
results = pd.DataFrame({
    "Cantidad de Datos": [1000, 2000, 4000, 8000, 16000, 25000],
    "Promedio de Tiempos (ms)": [mean_1000, mean_2000, mean_4000, mean_8000, mean_16000, mean_25000]
})

# Guardar el DataFrame en un archivo CSV
results.to_csv("/Users/edgarchambilla/Desktop/UTEC_CODE/BD/PROYECTO-BD2/docs/data.csv", index=False)

results
```

```python
# Graficar los datos promediados
plt.figure(figsize=(10, 6))
plt.plot(results["Cantidad de Datos"], results["Promedio de Tiempos (ms)"], marker='o')

plt.title("Promedio de Tiempos vs Cantidad de Datos")
plt.xlabel("Cantidad de Datos")
plt.ylabel("Promedio de Tiempo (ms)")
plt.grid(True)
plt.show()
```
